```
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df.groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort,ascending=False)
    .head(top)
  ).plot(kind='bar', figsize=(18,10), subplots=subplots)
```

```
group_and_barplot('param_kraj-pochodzenia', feat_sort='size');
```

```
group_and_barplot('param_marka-pojazdu', feat_sort='mean');
```

```
group_and_barplot('param_color', feat_sort='median');
```

```
group_and_barplot('param_pojemność-skokowa', feat_sort='median');
```
